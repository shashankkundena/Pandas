```python
import pandas as pd 
```

```python
df=pd.read_excel(r"C:\Users\shash\Downloads\Customer Call List.xlsx")
df
```

```python
#Removing duplicates
df = df.drop_duplicates()
df
```

```python
#Drop Unnecessary Columns

df = df.drop(columns='Not_Useful_Column')
df
```

```python
#Cleaning Last_Name using strip function

df['Last_Name'] = df['Last_Name'].str.strip('123./_')
```

```python
df
```

```python
#Formatting the Phone Number
```

```python
#converting everything into string
df['Phone_Number'] = df['Phone_Number'].astype(str)
```

```python
#Removing any non-numericals 

df['Phone_Number'] = df['Phone_Number'].replace('[^0-9]','' ,regex=True)
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:x[0:3] + '-' + x[3:6] + '-' + x[6:])
df
```

```python
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('--','')
```

```python
df
```

```python
#Cleaning the address using split 

df[['Street Address', 'State', 'ZipCode']] = df['Address'].str.split(',',n=2, expand=True)
df
```

```python
#Replaces Y and N with Yes and No
df['Paying Customer'] = df['Paying Customer'].replace({'Y':'Yes', 'N':'No', 'Yes':'Yes', 'No':'No'})
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Y':'Yes', 'Yes':'Yes', 'N':'No','No':'No'})
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('NaN','')
df
```

```python
#df.replace('N/a','')

df = df.fillna('')
df
```

```python
#Removing the Rows where Do_Not_Contact is Yes
```

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Yes':
        df.drop(x,inplace = True)

df
```

```python
#Removing the columns without Phone Number 
```

```python
for x in df.index: 
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)

df
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].replace('','No')
```
